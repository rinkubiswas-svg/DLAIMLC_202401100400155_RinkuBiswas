```python
X = [0.75, 0.65, 0.80]
W = [0.4, 0.3, 0.5]
b = -0.2
lr = 0.01
y = 1
```

```python
# Step 2: Net input (z)
z = (W[0]*X[0]) + (W[1]*X[1]) + (W[2]*X[2]) + b
print("Net input (z):", z)
```

```python
y_pred = 1 if z >= 0 else 0
print("Predicted output:", y_pred)
if(y_pred>0):

  print("Pass")
else:

  print("Fail")
  error = y - y_pred
  print("Error:", error)

  for i in range(len(W)):
    W[i]=W[i]+(lr*error*X[i])

  print("Updated weights: ",W)

  b=b+(lr*error)
  print("Updated bias: ",b)

  z_new = (W[0]*X[0]) + (W[1]*X[1]) + (W[2]*X[2]) + b
  y_pred_new = 1 if z_new >= 0 else 0

  print("New net input:", z_new)
  print("New prediction:", y_pred_new)

  if(y_pred_new<=0):
    print("Fail")
  else:
    print("Pass")
```
